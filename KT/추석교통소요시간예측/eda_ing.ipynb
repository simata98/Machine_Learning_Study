## 시간별 분석

```python
hour_2017_df = df_2017_busan.groupby('hour', as_index=False).mean()
hour_2018_df = df_2018_busan.groupby('hour', as_index=False).mean()
hour_2019_df = df_2019_busan.groupby('hour', as_index=False).mean()
hour_2020_df = df_2020_busan.groupby('hour', as_index=False).mean()
hour_2021_df = df_2021_busan.groupby('hour', as_index=False).mean()
```

```python
# f, ax = plt.subplots(5, 1, figsize=(16, 8))
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 1, 1) 
sns.lineplot(x='hour', y='spend_time',data=hour_2017_df ,label='2017')
sns.lineplot(x='hour', y='spend_time',data=hour_2018_df ,label='2018')
sns.lineplot(x='hour', y='spend_time',data=hour_2019_df ,label='2019')
sns.lineplot(x='hour', y='spend_time',data=hour_2020_df ,label='2020')
sns.lineplot(x='hour', y='spend_time',data=hour_2021_df ,label='2021')

plt.show()
```

2017년의 경우 밤이 깊어질 수록 오히려 더 오래걸리는 것을 볼 수 있는데, 2017년 10월의 경우 추석이 월요일 하루 빼고 전부다 휴일이었다. 2일 하루를 휴일을 사용한다면 10일 정도를 쉴 수 있었다.  
http://www.ohmynews.com/NWS_Web/View/at_pg.aspx?CNTN_CD=A0002362954  
위 기사에 따르면 추석기간이 길 경우, 귀성객이 분산이 되어 소요시간이 줄어들 것처럼 보이나 사실 여느 나들이철보다는 훨씬 많은 이용객이 발생하여 막힐 일이 없던 곳에서 꽉 막히는 경우가 발생한다고 한다.

이 그래프를 보고 느낀점은 코로나 발발(2020년) 후 귀성객이 이전 3년보다 많이 감소한 것을 볼 수 있다. 특히 2020년에 가장 적게 귀성길에 올랐다.  
만약 예측을 해야한다면 2020년 그래프는 코로나 발발이라는 특수한 경우로 인해 다른 그래프 분포와 상이하므로 제외하는 것도 고려해볼만 하다.

```python
hour_mean_df = pd.DataFrame(columns=['spend_time_mean'])
hour_mean_df = pd.concat([hour_mean_df, hour_2017_df['hour']], axis=1)
```

```python
hour_mean_df['spend_time_mean'] = (hour_2017_df['spend_time'] + hour_2018_df['spend_time'] + hour_2019_df['spend_time'] + hour_2020_df['spend_time'] + hour_2021_df['spend_time']) / 5
```

```python
# f, ax = plt.subplots(5, 1, figsize=(16, 8))
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 1, 1) 
sns.lineplot(x='hour', y='spend_time_mean',data=hour_mean_df ,label='spend_time_mean')
plt.show()
```

각 연도별 spend_time을 평균을 내어 시간별로 나타내었을 때 부산으로 귀성하는데 걸리는 시간은 오전 4시 부근이 가장 적게 걸렸다는 것을 알 수 있다.  
즉, 부산으로 빨리 가기 위해서는 이른 새벽에 출발하는 것이 좋다.

## 날씨 추가를 위한 전처리

```python
weather_2017 = pd.read_csv('./data/데이터셋/2017.csv', encoding='cp949')
weather_2018 = pd.read_csv('./data/데이터셋/2018.csv', encoding='cp949')
weather_2019 = pd.read_csv('./data/데이터셋/2019.csv', encoding='cp949')
weather_2020 = pd.read_csv('./data/데이터셋/2020.csv', encoding='cp949')
weather_2021 = pd.read_csv('./data/데이터셋/2021.csv', encoding='cp949')
```

```python
original_time_df = pd.DataFrame(columns=['original_time'])
merge_list = [original_time_df, weather_2017['일시'][:168], weather_2018['일시'][:168], weather_2019['일시'][:168], weather_2020['일시'][:168], weather_2021['일시'][:168]]
original_time_df = pd.concat(merge_list, ignore_index=True, axis=0)
```

```python
original_time_df.drop(columns=['original_time'], inplace=True)
original_time_df.columns = ['original_time']
```

```python
df_busan = pd.concat([df_busan, original_time_df['original_time']], axis=1)
```

```python
df_busan.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
seoul_weather_df = pd.concat([weather_2017.loc[:167], weather_2018.loc[:167], weather_2019.loc[:167], weather_2020.loc[:167], weather_2021.loc[:167]], axis=0)
busan_weather_df = pd.concat([weather_2017.loc[168:], weather_2018.loc[168:], weather_2019.loc[168:], weather_2020.loc[168:], weather_2021.loc[168:]], axis=0)
```

```python
drop_weather_columns = ['지점', '지점명', '지면상태(지면상태코드)']
seoul_weather_df.drop(columns=drop_weather_columns, inplace=True)
busan_weather_df.drop(columns=drop_weather_columns, inplace=True)
```

```python
seoul_weather_df.fillna(0, inplace=True)
busan_weather_df.fillna(0, inplace=True)
```

```python
for col_name in seoul_weather_df.columns:
  seoul_weather_df.rename(columns={col_name : '서울'+col_name}, inplace=True)
```

```python
for col_name in busan_weather_df.columns:
  busan_weather_df.rename(columns={col_name : '부산'+col_name}, inplace=True)
```

```python
seoul_weather_df.rename(columns={'서울일시':'original_time'}, inplace=True)
busan_weather_df.rename(columns={'부산일시':'original_time'}, inplace=True)
```

```python
seoul_weather_df.to_csv('./data/데이터셋/seoul_weather_전처리.csv')
busan_weather_df.to_csv('./data/데이터셋/busan_weather_전처리.csv')
```
