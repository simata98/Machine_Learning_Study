```python
import pandas as pd
from pycaret.classification import *
from time import time
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
train = pd.read_csv('/content/drive/MyDrive/capstone/UNSW_NB15/UNSW_NB15_training-set.csv')
test = pd.read_csv('/content/drive/MyDrive/capstone/UNSW_NB15/UNSW_NB15_testing-set.csv')
train.shape, test.shape
```

```python
train.info()
```

```python
drop_columns = ['id', 'attack_cat']
train.drop(columns=drop_columns, inplace=True)
test.drop(columns=drop_columns, inplace=True)
```

```python
clf = setup(data = train, target='label', silent=True, fix_imbalance=True, n_jobs=-1, categorical_features=['proto', 'service', 'state'])
```

```python
top5_models = compare_models(fold = 5, round = 3, sort = 'f1', n_select = 5)
```

```python
top5_models
```

```python
blended = blend_models(estimator_list = top5_models, fold = 5, method = 'soft')
```

```python
pred_holdout = predict_model(blended)
```

```python
final_model = finalize_model(blended)
```

```python
predictions = predict_model(final_model, data = test)
```
