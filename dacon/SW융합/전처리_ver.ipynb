```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 불러오기
train  = pd.read_csv('./data/train.csv')
test  = pd.read_csv('./data/test.csv')

pd.set_option('display.max_columns', None)
```

```python
train.head()
```

### 중복 행 제거

```python
train.drop_duplicates(inplace=True, keep='last')
```

### 결측치 보간

```python
# Education
'''
education의 경우 null 값을 평균값인 2.0으로 채워넣는다
'''
train[train['education'].isnull()]  = train[train['education'].isnull()].fillna(2.0)
```

```python
# Urban
'''
urban의 경우 결측치가 없음
'''
```

```python
# gender
'''
평균값인 2로 채워 넣는다.
'''
train[train['urban'].isnull()] = train[train['urban'].isnull()].fillna(2.0)
```

```python
# engnat
'''
많이 분포되어 있는 1.0으로 채워 넣는다.
'''
train[train['engnat'].isnull()] = train[train['engnat'].isnull()].fillna(1.0)
```

```python
# age
'''
결측치 없음.
'''
```

```python
# hand
'''
가장 많은 손이 분포되어 있는 1.0으로 채워넣는다.
'''
train[train['hand'].isnull()] = train[train['hand'].isnull()].fillna(1.0)
```

```python
# religion
'''
가장 많이 분포하고 있는 2로 채워넣는다
'''
train[train['religion'].isnull()] = train[train['religion'].isnull()].fillna(2.0)
```

```python
# orientation
'''
결측치 모두 1로 변환
'''
train[train['orientation'].isnull()] = train[train['orientation'].isnull()].fillna(2.0)
```

```python
# voted
'''
결측치 모두 2로 변환
'''
train[train['voted'].isnull()] = train[train['voted'].isnull()].fillna(2.0)
```

```python
# married
'''
결측치 모두 1로 변환
'''
train[train['married'].isnull()] = train[train['married'].isnull()].fillna(1.0)
```

```python
# familysize
'''
결측치 모두 2로 변환
'''
train[train['familysize'].isnull()] = train[train['familysize'].isnull()].fillna(2.0)
```

```python
# ASD
'''
결측치 모두 2로 변환
'''
train[train['ASD'].isnull()] = train[train['ASD'].isnull()].fillna(2.0)
```

```python
# 설문조사 항목 결측치 처리
# 앞 순번 설문조사한 사람의 숫자를 채워넣기로 함
train = train.fillna(method='ffill')
```

### 이상치 제거

```python
# 수치형 데이터 분리
numeric_feature = train.columns[(train.dtypes=='int64') | (train.dtypes=='float64')]
num_train = train[numeric_feature]

# 박스플롯
fig, axes = plt.subplots(7, 10, figsize=(25, 20))


fig.suptitle('feature distributions per quality', fontsize= 40)
for ax, col in zip(axes.flat, num_train.columns[:-1]):
    sns.boxplot(x='nerdiness', y= col, ax=ax, data=num_train)
    ax.set_title(col, fontsize=20)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

설문조사의 경우 이상치라도 의미있는 결과가 있을 것이라고 생각하여 따로 건들지 않기로 함  

```python
# 설문시간이 너무 길어서 비정상적인 설문일 수도 있을 것이라 생각, -> drop
intro_idx = train[train['introelapse'] > 400000].index
train.drop(intro_idx, inplace=True)

test_idx = train[train['testelapse'] > 200000].index
train.drop(test_idx, inplace=True)

surv_idx = train[train['surveyelapse'] > 150000].index
train.drop(surv_idx, inplace=True)

train.reset_index(inplace=True, drop=True)
```

```python
# age, familysize가 지나치게 큰 row 삭제
age_idx = train[train['age'] > 100].index
train.drop(intro_idx, inplace=True)

fam_idx = train[train['familysize'] > 20].index
train.drop(fam_idx, inplace=True)

train.reset_index(inplace=True, drop=True)
```

### 학습

```python
X = train.drop(labels=['index'], axis=1)
y = train['nerdiness']
```

```python
from pycaret.classification import *

model = setup(
    data = X,
    target = "nerdiness",
    use_gpu = True,
    train_size=0.95,
    session_id = 10
)
```

```python
top_5_model = compare_models(
                sort="AUC",
                n_select=5, 
                exclude=['xgboost'])
```

```python
blended = blend_models(estimator_list = top_5_model, fold = 5, method = 'soft')
```

```python
pred_holdout = predict_model(blended)
```

```python
final_model = finalize_model(blended)
```

```python
test.drop(labels=['index'], axis=1, inplace=True)
```

```python
predictions = predict_model(final_model, data = test)
```
